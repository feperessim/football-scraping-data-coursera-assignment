```python
import pandas as pd
import requests

from bs4 import BeautifulSoup
```

<h1>Santos FC Team</h1> 

```python
santos_fc_url = 'https://en.wikipedia.org/wiki/List_of_Santos_FC_seasons'
table_class = 'wikitable'
```

<h2>Checking Response status</h2>

```python
response=requests.get(santos_fc_url)
print(response.status_code)
```

<h2>Parsing Santos FC Data</h2>

```python
soup = BeautifulSoup(response.text, 'html.parser')
santos_fc_table = soup.findAll('table',{'class':"wikitable"})
```

```python
df=pd.read_html(str(santos_fc_table))
# convert list to dataframe
# since there are many tables using
# the same class, the table we want is
# at position 2
df=pd.DataFrame(df[2])
df.head()
```

```python
df['Domestic leagues[7][8][9][24]']
```

```python
df['Season']
```

<h2>Concat the years with the championships data<h2>

```python
santos_fc_df = df[['Season', 'Domestic leagues[7][8][9][24]']]
santos_fc_df.columns = santos_fc_df.columns.droplevel()
santos_fc_df
```

<h2>Filtering Data for further use</h2>

```python
santos_fc_df['Season'] = santos_fc_df['Season'].apply(lambda year: year.strip()[:4])
santos_fc_df = santos_fc_df.astype({"Season": 'int32'}, errors='raise')
santos_fc_df = santos_fc_df[(santos_fc_df['Season'] >= 1980) & (santos_fc_df['Season'] < 2022)]
santos_fc_df = santos_fc_df[santos_fc_df['Tournament'] == 'Brasileirão']
```

```python
santos_fc_df.head()
```

```python
santos_fc_df.tail()
```

```python
santos_fc_df.info()
```

```python
santos_fc_df.to_csv('santos_fc_list_of_seasons_df.csv', index=False)
```
